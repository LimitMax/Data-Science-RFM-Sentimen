# Load Data & Integration

tahap ini dilakukan untuk membaca data dan menggabungkan data di setiap tabel

```python
pd.set_option('display.max_columns', None)

missng_data = ['Nan', 'NaN', 'na', 'nan', 'NA', 'N/A', 'n/a', 'missing', 'MISSING', '']

order = pd.read_csv('orders_dataset.csv', na_values=missng_data)
order_items = pd.read_csv('order_items_dataset.csv', na_values=missng_data)
customer = pd.read_csv('customers_dataset.csv', na_values=missng_data)
geoloc = pd.read_csv('geolocation_dataset.csv', na_values=missng_data)
product = pd.read_csv('products_dataset.csv', na_values=missng_data)
payment = pd.read_csv('order_payments_dataset.csv', na_values=missng_data)
seller = pd.read_csv('sellers_dataset.csv', na_values=missng_data)
```

```python
#integrasi data dengan inner join
data = pd.merge(order, order_items, on='order_id', how='inner')
data = pd.merge(data, customer, on='customer_id', how='inner')

data.head()
```

```python
#hapus kolom customer_state,order_delivered_carrier_date,order_estimated_delivery_date,shipping_limit_date
data = data.drop(columns=['customer_state', 'shipping_limit_date'])

data.head()
```

```python
#cek nilai unik order_status
data['order_status'].unique()
```

```python
data = pd.merge(data, product, on='product_id', how='inner')

data.head()
```

```python
#hapus kolom product_name_length,product_description_length,product_photos_qty,product_length_cm,product_height_cm,product_width_cm
data = data.drop(columns=['product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_length_cm', 'product_height_cm', 'product_width_cm'])

data.head()
```

```python
#integrasi dengan payment
data = pd.merge(data, payment, on='order_id', how='inner')

data.head()
```

```python
#integrasi dengan seller
data = pd.merge(data, seller, on='seller_id', how='inner')

data.head()
```

```python
#hapus kolom seller_zip_code_prefix, seller_city
data = data.drop(columns=['seller_zip_code_prefix', 'seller_state'])
```
